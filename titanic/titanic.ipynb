```python
# import matplotlib.pyplot as plt

# Analysis 1: Survival Rate Analysis
survival_counts = df['Survived'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(survival_counts, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
plt.title('Survival Rate')
plt.show()

# Analysis 2: Passenger Class Analysis
class_survival = df.groupby('Pclass')['Survived'].mean()
plt.figure(figsize=(8, 6))
class_survival.plot(kind='bar', color='skyblue')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.show()

# Analysis 3: Age Distribution
plt.figure(figsize=(8, 6))
df['Age'].hist(bins=20, color='salmon', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Analysis 4: Gender Analysis
gender_survival = df.groupby('Sex')['Survived'].mean()
plt.figure(figsize=(6, 6))
gender_survival.plot(kind='bar', color=['lightblue', 'pink'])
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Drop rows with missing values in 'Age' or 'Fare' columns
df_3d = df.dropna(subset=['Age', 'Fare'])

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot data points for passengers who survived
ax.scatter(df_3d[df_3d['Survived'] == 1]['Age'],
           df_3d[df_3d['Survived'] == 1]['Fare'],
           df_3d[df_3d['Survived'] == 1]['Survived'],
           c='g', marker='o', label='Survived')

# Plot data points for passengers who did not survive
ax.scatter(df_3d[df_3d['Survived'] == 0]['Age'],
           df_3d[df_3d['Survived'] == 0]['Fare'],
           df_3d[df_3d['Survived'] == 0]['Survived'],
           c='r', marker='x', label='Not Survived')

ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_zlabel('Survived')
ax.set_title('3D Analysis of Age, Fare, and Survival Status')
ax.legend(loc='upper right')

plt.show()
```
