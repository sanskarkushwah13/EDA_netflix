```python
""" Q. 1) For 'House of Cards', what is the Show Id and Who is the Director of this show ? """ 

df.columns
```

```python
# v=df[df.show_id=="House of Cards"].values
# v1=df[df['director']=='Blood & Water']
# v
# v1
house_of_cards = df[df['title'] == 'House of Cards']
if not house_of_cards.empty:
    show_id = house_of_cards['show_id'].values[0]
    director = house_of_cards['director'].values[0]
    print("Show ID: ", show_id)
    print("Director: ", director)
else:
    print("No information found for 'House of Cards'.")
```

```python
""" Q. 2) In which year the highest number of TV Shows & Movies were released? Show with 
Bar Graph."""

show =df.release_year.value_counts()
plt.figure(figsize=(10, 6))
show.plot(kind='bar', color='blue')
plt.xlabel('Year')
plt.ylabel('Number of TV Shows & Movies')
plt.title('Number of TV Shows & Movies Released per Year')
# plt.xticks(rotation=45)
# plt.axvline(x=show[:5], color='red', linestyle='--', linewidth=2,label='Year with Highest Releases: {}'.format(show))
plt.legend()
plt.show()
```

```python
#  Q. 10) What is the maximum duration of a Movie/Show on Netflix ?""

df.columns

df.duration.value_counts()
movies_and_shows = df[df['type'].isin(['Movie', 'TV Show'])]

movies_and_shows['duration'] = movies_and_shows['duration'].str.extract('(\d+)')
movies_and_shows['duration'] = pd.to_numeric(movies_and_shows['duration'])
max_duration = movies_and_shows['duration'].max()
max_duration
```
